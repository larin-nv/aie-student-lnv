В данном случае новый массив arr_slice_vector4 берет срез от индекса 6 включительно до конца массива

Векторные операции и базовые статистики

```python
source_array = np.arange(8)
source_array
```

Операция возведения в степень поэлементно

```python
exponented_array = source_array ** 2
exponented_array
```

Операция сложения массивов

```python
sum_array = source_array + exponented_array
sum_array
```

В данном примере происходит сложение элементов в соответствии с индексами, то есть, элемент одного массива с индексом 0 суммируется с элементом другого массива с индексом 0 и далее. Таким образом, получается новый массив

```python
diff_array = source_array - exponented_array
diff_array
```

По тому же пути происходит вычитание одного массива из другого

Абсолютно также работает и операция умножения массивов

```python
mult_array = source_array * exponented_array
mult_array
```

Для того, чтобы найти среднее значение элемента массива, в NumPy предусмотрен метод mean

```python
np.mean(source_array)
```

Для того, чтобы найти сумму всех элементов массива, в NumPy предусмотрен метод sum

```python
np.sum(source_array)
```

Или

```python
source_array.sum()
```

```python
source_array.min()
```

Пример на работу с NumPy

Есть данные о времени подготовки студентов к экзамену (x) и то, сколько баллов они получат (y)

```python
import math
```

```python
x_i = np.array([2, 4, 6, 8, 10])
y_i = np.array([55, 65, 75, 85, 95])
```

```python
x_i
```

```python
y_i
```

Возьмем средние значения

```python
Xch = np.mean(x_i)
Ych = np.mean(y_i)
```

```python
Xch
```

```python
Ych
```

Теперь возьмем выборочные дисперсии

```python
Sx2 = np.sum((x_i - Xch) ** 2) / (len(x_i) - 1)
Sy2 = np.sum((y_i - Ych) ** 2) / (len(y_i) - 1)
```

```python
Sx2
```

```python
Sy2
```

Найдем средне квадратичные отклонения

```python
Sx = math.sqrt(Sx2)
Sy = math.sqrt(Sy2)
```

```python
Sx
```

```python
Sy
```

Ищем ковариацию, чтобы узнать степень линейной взаимосвязи

```python
cov_xy = np.sum((x_i - Xch) * (y_i - Ych)) / (len(x_i) - 1)
cov_xy
```

Коэффициент корелляции Пирсона

```python
r_xy = cov_xy / (Sx * Sy)
r_xy
```

Это значит полную положительную линейную связь между временем подготовки и количествомм баллов на экзамене

Чем больше часов подготовки, тем выше баллы на экзамене
